### Splitting the dataset

```python
X = data.drop(columns='W')
y = data['W']
y.head(10)
```

```python
X
```

### Feature selection

as we hve mentioned, the dataset is quite small, so there will be problem of overfitted model, so we will reduce the some of the features in the data set 

```python
# we will select best features using chi square technique

from sklearn.feature_selection import SelectPercentile
from sklearn.feature_selection import chi2

# selecting best 80 percent features
s_percentile = SelectPercentile(score_func=chi2, percentile=80)
s_percentile = s_percentile.fit(X,y)

col = s_percentile.get_support(indices=True)
features = X.columns[col]
```

```python
print(features)
print(col)
```

```python
X = data[features]
```

```python
# Now lets check the checking the distribution of data
X.plot(kind='density',sharex=False,subplots=True,layout=(6,3),figsize=(20,15))
plt.show()

X.skew().sort_values(ascending=False)
```

most of the data looks normally dstributed except few whose have little bit of skewness

so I will be removing the skewness from those data using suare root transformation

```python
X.R = np.cbrt(X.R)
X.E = np.cbrt(X.E)
X.CG = np.cbrt(X.CG)
X.H = np.cbrt(X.H)
X.SV = np.cbrt(X.SV)
X.HR = np.cbrt(X.HR)

X
```

```python
# Lets begin with checking the distribution of data

X.plot(kind='density',sharex=False,subplots=True,layout=(6,3),figsize=(20,15))
plt.show()

X.skew().sort_values(ascending=False)
```

Although it seems that there are still some skewnes in som of the column, so to confirm lets plot it in box plot to see if we still needs to treat the skewness

```python
from scipy.stats import zscore
zscore(X).plot(kind='box',sharex=False,subplots=True,layout=(6,3),figsize=(20,15))
plt.show()

X.skew().sort_values(ascending=False)
```

as per the threshold in zscore (-3 to +3). it is confirmed from the box plot above that all the dataset  falls under this threshold so we dnt need to treat it further, so we can go head and scale the data

### Scaling the data

```python
from sklearn.preprocessing import StandardScaler

x_scaled = StandardScaler().fit_transform(X)
x_scaled
```

### Training the data

As we have very small dataset I will only be using simple model (linear regression) to predict the data, because choosing complex may lead to overfitted model

    - linear regression

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, mean_absolute_error
# best the best random state for linear regrssion model
ri = 0
mae = 10
score = 0

for i in range(0,200):
    x_train, x_test, y_train, y_test = train_test_split(x_scaled,y,train_size=.75,random_state=i)
    lm = LinearRegression()
    lm.fit(x_train,y_train)
    y_pred = lm.predict(x_test)
    pri = i
    pmae = mean_absolute_error(y_test,y_pred)
    pscore = lm.score(x_test,y_test)
    
    if mae > pmae:
        mae = pmae
        ri = pri
        score = pscore
print('mean absolute error - ',mae,'\n','random state -',ri,'\n','Score --',pscore)
```
